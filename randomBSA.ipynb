Proprtion of shots made

```python
import re
import operator
import pandas as pd

# Regex expression for extracting shot information
p = re.compile('(\s{2}|\' )([\w+ ]*)')

# Filter for events related to shot attempts (EVENTMSGTYPE 1 for make, 2 for miss)
shot_attempts = df[(df['EVENTMSGTYPE'] == 1) | (df['EVENTMSGTYPE'] == 2)]

# Dictionary to store shot attempts categorized by type and result
shot_results = {}

# Loop over shot attempts
for index, play in shot_attempts.iterrows():
    description = play['HOMEDESCRIPTION'] if pd.notna(play['HOMEDESCRIPTION']) else play['VISITORDESCRIPTION']
    if pd.notna(description):  # Check if description is not null
        # Check if description is a string
        # Extract action from description using regex
        match = p.search(description)
        if match:
            event_msg_action = re.sub(' ', '_', match.groups()[1].rstrip()).upper()
             # Classify as make or miss
            if play['EVENTMSGTYPE'] == 1:
                result = 'MAKE'
            else:
                result = 'MISS'
             # Add shot attempt to dictionary
            if event_msg_action not in shot_results:
                shot_results[event_msg_action] = {'MAKE': 0, 'MISS': 0}
            shot_results[event_msg_action][result] += 1

# Sort shot types alphabetically
sorted_shot_results = sorted(shot_results.items(), key=operator.itemgetter(0))

# Output the results
for shot_type, result in sorted_shot_results:
    print(f"{shot_type}:")
    print(f"\tMakes: {result['MAKE']}")
    print(f"\tMisses: {result['MISS']}")
    ##print(f"\tTotal: {result['MISS'] + result['MAKE']}")
```
